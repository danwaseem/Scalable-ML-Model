```python
import pandas as pd
import numpy as np
import sklearn.model_selection as model_selection
from sklearn.covariance import EllipticEnvelope
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
avg=0
```

## Ethereuem Address 1 
### 0x0067f95a79c3c404a9d128168ddfdf3cb70c0852

```python
dfs = pd.read_csv(r"D:\Desktop\nitk\6th_SEM\ML_Project\transaction_data\1_0x0067f95a79c3c404a9d128168ddfdf3cb70c0852.csv")
dfs1=dfs
display(dfs)
```

### Pre-Processing

```python
dfs=dfs.drop(['Txhash','Blockno','UnixTimestamp','DateTime','From','To'],axis=1)
dfs=dfs.dropna()
display(dfs)
```

### Splitting the data

```python
train, test = model_selection.train_test_split(dfs,test_size=0.25, random_state=101)
print("TESTING DATA")
display(train)
print("TRAINING DATA")
display(test)
```

### Training and testing the model

```python
model = IsolationForest(contamination=0.01,random_state=101)
model.fit(train)

y_pred1=model.predict(test)
acc=(y_pred1==-1).sum()/test.shape[0]
print("The Accuracy of the anamoly detection unit (Isolation forest) is", (1-acc)*100,"%")
```

```python
model=OneClassSVM(nu=0.01)
model.fit(train)

y_pred2=model.predict(test)
acc=(y_pred2==-1).sum()/test.shape[0]
print("The Accuracy of the the anamoly detection (One Class SVM) unit is ", (1-acc)*100,"%")
```

```python
model = EllipticEnvelope(contamination=0.02)
model.fit(train)

y_pred3=model.predict(test)
acc=(y_pred3==-1).sum()/test.shape[0]
print("The Accuracy of the the anamoly detection (Minimum Coeriance Determinant) unit is ", (1-acc)*100,"%")
```

```python
print("Isolation factor has an higher accuracy than the others, so we will use Isolation forest for the remaining user transaction.")
```

```python
model = IsolationForest(contamination=0.01,random_state=101)
train['anamoly']=model.fit_predict(train)
a=train.loc[train['anamoly']==-1]
test['anamoly']=model.predict(test)
```

### Transactions which has been marked as anamoly

```python
b=test.loc[test['anamoly']==-1]

c=pd.concat([a,b])
merged_left = pd.merge(left=dfs1, right=c, how='right',sort=True)
display(merged_left)
```
