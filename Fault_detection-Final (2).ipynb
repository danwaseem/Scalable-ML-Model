```python
import pandas as pd
from xgboost import XGBClassifier
import numpy as np
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
import sklearn.model_selection as model_selection
from sklearn import metrics   #Additional scklearn functions
from sklearn.model_selection import learning_curve, GridSearchCV  #Perforing grid search
from sklearn.metrics import confusion_matrix, accuracy_score  
from sklearn.metrics import classification_report
import matplotlib.pylab as plt
%matplotlib inline
```

```python
dfs = pd.read_csv(r'D:\Desktop\nitk\6th_SEM\ML_Project\dataset.csv')
dfs
```

```python
dfs.columns = [c.replace(' ', '_') for c in dfs.columns]
dfs
```

```python
dfs=dfs.drop('Index',axis=1)
```

```python
dfs=dfs.drop(['Address','ERC20_most_rec_token_type'],axis=1)
```

```python
dfs = dfs.loc[:, (dfs != 0).any(axis=0)]
```

```python
dfs=dfs.drop(['ERC20_avg_time_between_sent_tnx','ERC20_avg_time_between_rec_tnx','ERC20_avg_time_between_rec_2_tnx','ERC20_avg_time_between_contract_tnx','ERC20_min_val_sent_contract','ERC20_max_val_sent_contract','ERC20_avg_val_sent_contract'],axis=1)
```

```python
dfs=dfs.drop(['ERC20_most_sent_token_type'],axis=1)
```

```python
dfs.fillna(0,inplace=True)
```

```python

from sklearn import preprocessing

x = dfs.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
dfs=pd.DataFrame(x_scaled, columns=dfs.columns)
```

# After Pre-Processing

```python
dfs
```

```python
X=dfs.drop('FLAG',axis=1)
X
```

```python
y=dfs['FLAG']
y
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y,test_size=0.25, random_state=101)
```

# After Splitting with 0.75 ratio

# SVM Classifier

```python
from sklearn.svm import SVC
```

```python
SVM2=SVC(C=80,gamma=190)
SVM2.fit(X_train, y_train)
y_pred2 = SVM2.predict(X_test)
acc=(y_test != y_pred2).sum()/X_test.shape[0]
print("The Accuracy of the following image processing unit is ", (1-acc)*100,"%")
```

### Confusion Matrix

```python
cm= confusion_matrix(y_test, y_pred2)  
ac = accuracy_score(y_test, y_pred2)
print(cm)
print("Accuracy:", ac * 100, "%")
```

### Classification Report



```python
results = classification_report(y_test, y_pred2)
print (results)
```

# XGBoost Classifier

```python
xgb3 = XGBClassifier(
 learning_rate =0.1,
 n_estimators=1000,
 max_depth=5,
 min_child_weight=1,
 gamma=0.1,
 subsample=0.7,
 colsample_bytree=0.6,
 reg_alpha=0.001,
 nthread=4,
 scale_pos_weight=1,
 seed=27)
```
